# one hot encoding

```python
Location = OneHotEncoder(sparse_output= False)
WindGustDir = OneHotEncoder(sparse_output= False)
WindDir9am =  OneHotEncoder(sparse_output= False)
WindDir3pm =  OneHotEncoder(sparse_output= False)
```

```python
l_ohe= Location.fit_transform(df[['Location']])
wind_ohe= WindGustDir.fit_transform(df[['WindGustDir']])
dir_ohe= WindDir9am.fit_transform(df[['WindDir9am']])
dir2_ohe= WindDir3pm.fit_transform(df[['WindDir3pm']])
```

```python
df_loc= pd.DataFrame(l_ohe, columns= Location.categories_)
df_loc= df_loc.iloc[:,1:]
df_loc
```

```python
df_wind= pd.DataFrame(wind_ohe, columns= WindGustDir.categories_)
df_wind= df_wind.iloc[:,1:]
df_wind
```

```python
df_dir= pd.DataFrame(dir_ohe, columns= WindDir9am.categories_)
df_dir= df_dir.iloc[:,1:]
df_dir
```

```python
df_dir2= pd.DataFrame(dir2_ohe, columns= WindDir3pm.categories_)
df_dir2= df_dir2.iloc[:,1:]
df_dir2
```

```python
X= df.drop(['Location', 'WindGustDir', 'WindDir9am', 'WindDir3pm', 'RainTomorrow'], axis= 1)
y= df['RainTomorrow']
```

```python
X= pd.concat([X, df_loc, df_wind, df_dir, df_dir2 ], axis=1)
```

```python
X.info()
```

```python
X= X.values
y= y.values
```

# voting classifier

```python
from sklearn.ensemble import VotingClassifier
```

```python
dtree=  DecisionTreeClassifier(max_depth= 10)
log_reg= LogisticRegression()
rand_for= RandomForestClassifier(max_depth= 10)
```

```python
est= [('dtree', dtree),
     ('log_reg', log_reg),
     ('rand_for',rand_for )]
```

```python
vc= VotingClassifier(estimators= est, voting= 'hard')
```

```python
vc.fit(X_train, y_train)
```

```python
vc_train_pred=vc.predict(X_train)
```

```python
accuracy_score(y_train,vc_train_pred )
```
